```python
#Qno6.Write a program using a for loop to print all twin primes between 1 and 100. (Twin primes: pairs of prime 
# numbers with a difference of 2, e.g., (3,5), (11,13)).

# Solution
n = 100
for num in range(2, n-1): 
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        for j in range(2, num+2):
            if (num+2) % j == 0:
                break
        else:
            print(num, num+2)

```

```python
# Qno 10.Write a program that accepts a number from the user and prints whether it is a Strong number (sum of 
# factorials of digits = number itself) using a for loop. Example: 145 => 1! + 4! + 5! = 145.
# solution
n = int(input("Enter a number"))
sum_fact = 0

for digit_char in str(n):       
    digit = int(digit_char)
    fact = 1
    for i in range(1,digit+1):  
        fact=fact*i
    sum_fact= sum_fact+fact

if sum_fact == n:
    print(n,"is a Strong number")
else:
    print(n,"is not a Strong number")
        
```

```python
# Qno.11 Write a program using a while loop to find the reverse of a number and check if the reversed number is 
# prime. Example: Input = 73 → Reverse = 37 → Prime.

# solution
n = int(input("Enter number"))
rev = 0
temp = n
while temp > 0:
    rev = rev*10 + temp%10
    temp //= 10
i = 2
prime = True
if rev < 2:
    prime = False
while i*i<=rev and prime:
    if rev%i == 0:
        prime = False
    i+=1

print("Reversed:", rev)
print("Prime" if prime else "Not prime")
```

```python
# QNO 13: Write a program using a while loop to check whether a number is a Duck number (a number containing zero 
# but not starting with zero, e.g., 202, 1203)
# Solution
num = input("Enter number:")
i = 1
duck = False
while i < len(num):
    if num[i] == '0':
        duck = True
        break
    i=i+ 1
print("Duck number" if duck else "Not a duck number")
```

```python
# QNO 14:Write a program using a while loop to accept a number and check if it is a Happy number. (A number is 
# happy if repeatedly replacing it with the sum of squares of its digits eventually reaches 1). Example: 19 is a 
# happy number.

# Solution
num = int(input("Enter number: "))
seen = set()
while num != 1 and num not in seen:
    seen.add(num)
    num = sum(int(d)**2 for d in str(num))
print("Happy" if num==1 else "Not happy")
```

```python
# QNO 19:Write a program using a while loop to accept a number and check whether it is a Kaprekar number. 
# (Kaprekar number: if square of the number can be split into two parts whose sum equals the number. 
# Example: 45²=2025 => 20+25=45).

# solution:
n = int(input("Enter number:"))
sq = str(n*n)
kap = False
for i in range(1, len(sq)):
    if int(sq[:i] or 0) + int(sq[i:] or 0) == n:
        kap = True
        break
if n==1 or kap:
    print("Kaprekar number")
else:
    print("Not Kaprekar")
```

```python
# QNO 20:Write a program to simulate an ATM machine using a while loop where a user can:
# • Check balance
# • Deposit money
# • Withdraw money (only if balance is sufficient)
# • Exit
# Continue until the user chooses to exit.

# solution:
balance = 1000
while True:
    print("1.Balance 2.Deposit 3.Withdraw 4.Exit")
    choice = int(input("Choice: "))
    if choice==1:
        print("Balance:", balance)
    elif choice==2:
        balance += int(input("Deposit amount: "))
    elif choice==3:
        amt = int(input("Withdraw amount: "))
        if amt <= balance: balance -= amt
        else: print("Insufficient balance")
    elif choice==4:
        print("Exit"); break
```
